```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
df = pd.DataFrame(columns = ['brand_id','price','link','title','brand_name'])


for page in range(1,100,1):
    source=requests.get(f'https://www.gittigidiyor.com/cep-telefonu?sra=hpd&sf={page}').text
    soup=BeautifulSoup(source,'lxml')
    
    for article in soup.find_all('article'):
#         print(article)
        m={
            "brand_id":article.find('footer')['id'].split('-')[1],
            "price":article.find('span',class_='buy-price').text.split()[0],
            "link":article.a['href'],
            "title":article.find('h3')['title'],
            "brand_name":article.find('h3')['title'].split(" ")[0],
           
           }
    
#         print(m)
        df=df.append(m,ignore_index=True)
        

          
        
        
         
        
        
#         df.append({'brand_id':brand_id,'price':price,'link':link,'title':title,'brand_name':brand_name},\
# article.find('span',class_='buy-price').text              ignore_index=True)
```

```python
df['price'].replace('\.','', regex=True,inplace=True)
```

```python
df['price']=df['price'].str.split(',').str[0]
```

```python
df["price"]=df["price"].astype(str).astype(int)
```

```python
df["brand_id"]=df["brand_id"].astype(str).astype(int)
```
